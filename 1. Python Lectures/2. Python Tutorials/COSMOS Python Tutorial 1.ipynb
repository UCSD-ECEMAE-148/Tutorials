### Printing in Python:
  <br>&emsp;Printing the output is a very important operation to learn any programming language as it helps greatly in the greatly in the process of debugging a code. `print()` is the function that is used in python for printing stuff. The `argument` for a print function can be any string or any object.The object will be converted into a string before written to the screen. Generally. we only input the required object to the `print()` function. However, the syntax of `print()` in python goes like: <br>
     &emsp;&emsp;**print(*objects, sep=' ', end='\n')** <br>

`sep` stands for the seperator that is to be used while printing multiple objects. <br>
`end` stands for the end input that is to be placed after printing all the input arguments
Few examples of using print function
     

```python
#Providing the input directly and the default values of sep is a whitespace and end is \n (a new line)

print("Welcome to COSMOS")
print('*********************')
#multiple inputs and multiple executions

print("1)Welcome to COSMOS","Autonomous Vehicles")
print("2)Intro to Python")
print('*********************')
#providing seperator argument

print("1)Welcome to COSMOS","Autonomous Vehicles",sep='\t')
print("2)Intro to Python","Excited!",sep='\t')

print('*********************')

#seperator and end arguments


print("1)Welcome to COSMOS","Nice to meet you",sep='space',end='\n')
print("2)Intro to Python","Excited!",sep='\t',end='\t')

print('*********************')
```

### Commenting : <br>

Pound key (#) is used for commenting a single line of code.  The same can be perfomed using `cmd` + `/` on Mac and `ctrl` + `/` on Windows. This is generally used for providing inline comments. <br>

For the purpose of multi line commenting, we generally provide the pound key(#) before every line of code. Another way of achieving multi line commenting is to use the `triple quote` i.e, `'''`. If a `'''` is used in body of the code python considers the entire block of the code following triple quote as comments until a closing `'''` is encountered.


### Indentation:<br>

Indentation refers to the spaces at the beginning of a code line. Where in other programming languages the indentation in code is for readability only, the indentation in Python is very important. Python uses indentation to indicate a block of code. All the lines of code with a particular level of indentation constitute a block of code. This will be clear once we reach loops and functions in the course.



```python
#Commenting using pound key

print('This is a basic way of commenting')

example = 'Another way of using the pound key' # Defining a variable string

print(example) #printing example variable

'''
 All the lines of code
 that are placed between triple
 quotes are considered to be comments
 in Python. This is typically used in function description
'''
print("Commenting blocks of code")
```

### Variables and Datatypes: <br>

When a program's running, everything the program needs ends up stored in the computer's memory. The program code itself will be stored in one area of memory, and the data it works on will also be stored somewhere in memory.
A variable is basically just a way to give a (meaningful) name to an area of memory, into which we can place certain values.<br>

There are a few rules for variable names:<br>
- Python variable names must begin with a letter (either upper or lower case) or an
underscore _ character.
- They can contain letters, numbers or underscore characters (but cannot begin with a number).
- Python variables are case sensitive.
- Variables are created when they are first attached to a value, using =.<br>

-     &emsp; & 	&emsp;AND	&emsp;Sets each bit to 1 if both bits are 1

-     &emsp;^	    &emsp;XOR	&emsp;Sets each bit to 1 if only one of two bits is 1
-     &emsp;~ 	&emsp;NOT	&emsp;Inverts all the bits

### Sequences in Python:
    
A sequence can be defined to be as an ordered set of items. Python has 3 major types of sequences:

- List - collection of objects and mutable
- Tuple - collection of objects and immutable

Examples:

```python
#Provide examples for different sequence types
string_1 = "Cluster11"
print(len(string_1))
list_1=["UCSD",4,3.1]
print(len(list_1))
tup_1 = ("UCSD",4,3.1)
print(len(tup_1))
```

#### Slicing:<br>

Slicing of a sequence means extracting a subset of the items in the collection following a pattern. Slicing typicall has [start : end : step] syntax where:
- start - represents the index position of the starting element
- end -  represents the index position of the **last element plus 1**. In simple words the right end of the bracket is not inclusive.
- step - The number of indices to incremented for every item



```python
string = "Autonomous Vehicles"
print("Lenght of the string is :",len(string))

print(string[0:6])
print((string[0:6:2]))

print(string[::-1])

#Negative slicing


#Additional examples on slicing using lists and tuples


```
