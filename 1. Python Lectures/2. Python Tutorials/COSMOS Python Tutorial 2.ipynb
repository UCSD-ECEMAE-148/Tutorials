#### Format Printing

str.format( )<br>

Syntax: { }.format(value)

Parameters:

value : Can be an integer, floating point numeric constant, string, characters or even variables.
Returntype: Returns a formatted string with the value passed as parameter in the placeholder position.

#### Lists in Python: <br>

Lists are used to store multiple items in a single variable. Lists are created using square brackets:
Example: `Fruits=['apple',2,30]`. List items are ordered, changeable, and allow duplicate values. The list is changeable, meaning that we can change, add, and remove items in a list after it has been created. <br>
All the items in a list need not be of the same data type. An item in a list can be an another list too. In such cases, it is referred as Nesting of lists. All sequences have some basic methods like `len`,

```python

Fruits=["apple","oranges","kiwi","grapes"]
print(len(Fruits))
print(Fruits[::-1])
```

```python
string_in = input("Enter a sentence:")
list_in  = string_in.split()
print(list_in)
```

```python
Fruits=["apple","oranges","kiwi","grapes"]
#Append

Fruits.append("cherries")
print(Fruits)

#Reverse
Fruits.reverse()
print('reverse',Fruits)
#sort
Fruits.sort()
print('sort',Fruits)
#sorted
Fruits_sorted = sorted(Fruits)
print('sorted',Fruits_sorted)
print('original',Fruits)
#insert

Fruits.insert(4,"Mangoes")

print(Fruits)
# Fruits[4]="Mangoes"
# Fruits.append("oranges")
# print(Fruits)

#remove
Fruits.remove("kiwi")
print(Fruits)

#pop
Fruits.pop(2)
print(Fruits)
```

```python
#Copying a list

list_1 = ["Apples","Oranges","Kiwi","Grapes","Cherries"]
list_2=list_1
list_3 = list_1.copy()
list_1.remove("Cherries")
print(list_1,list_2,list_3,sep='\n')
```

```python
#Concatenation of lists

list_5 = list_1+list_3
print(list_5)
```
